```python
# Realize a EDA (Exploratory Data Analysis) on the dataset of candidates 
import pandas as pd
from tabulate import tabulate

# Define the path to the dataset
path = r'C:\Users\btigr\Documents\UAO\5\ETL\ETL_2026_1\workshop_1\etl_workshop_01\raw\candidates.csv'

# Load the dataset
df = pd.read_csv(path, sep=';')

# Show the first few rows of the dataset
print('Showing first few rows of the dataset:')
print(tabulate(df.head(), headers='keys', tablefmt='psql'))
```

```python
# Explore type of data in each column
print('\nData types of each column:')
print(tabulate(df.dtypes.to_frame(), headers='keys', tablefmt='psql'))

# Know how many rows and columns the dataset has
print('\nNumber of rows and columns in the dataset:')
print(tabulate([df.shape], headers=['Rows', 'Columns'], tablefmt='psql'))
```

```python
# Know nulls and duplicates
print('\nNumber of null values in each column:')
print(tabulate(df.isnull().sum().to_frame(), headers='keys', tablefmt='psql'))

print('\nNumber of duplicate rows in the dataset:')
print(df.duplicated().sum())
```

```python
# Know countrys where most candidates aply
print('\nTop Countries where most candidates apply:')
print(tabulate(df['Country'].value_counts().head(10).to_frame(), headers='keys', tablefmt='psql'))

# Bar plot horizontal of top 10 countries where most candidates aply
import matplotlib.pyplot as plt

top_countries = df['Country'].value_counts().head(10)
plt.figure(figsize=(10, 6))
plt.barh(top_countries.index, top_countries.values)
plt.xlabel('Number of Candidates')
plt.title('Top 10 Countries where most candidates apply')
plt.gca().invert_yaxis()
plt.show()
```

```python
# Know technologies where most candidates aply
print('\nTop Technologies where most candidates apply:')
print(tabulate(df['Technology'].value_counts().head(10).to_frame(), headers='keys', tablefmt='psql'))

# Bar plot horizontal of top 10 technologies where most candidates aply
top_technologies = df['Technology'].value_counts().head(10)
plt.figure(figsize=(10, 6))
plt.barh(top_technologies.index, top_technologies.values)
plt.xlabel('Number of Candidates')
plt.title('Top 10 Technologies where most candidates apply')
plt.gca().invert_yaxis()
plt.show()
```

```python
# Know senority of candidates
print('\nTop Seniority levels of candidates:')
print(tabulate(df['Seniority'].value_counts().head(10).to_frame(), headers='keys', tablefmt='psql'))

# Bar plot horizontal of top 10 seniority levels of candidates
top_seniority = df['Seniority'].value_counts().head(10)
plt.figure(figsize=(10, 6))
plt.barh(top_seniority.index, top_seniority.values)
plt.xlabel('Number of Candidates')
plt.title('Top 10 Seniority levels of candidates')
plt.gca().invert_yaxis()
plt.show()
```

```python
# Know Years of experience of candidates 
print('\nTop Years of experience of candidates:')
print(tabulate(df['YOE'].value_counts().head(10).to_frame(), headers='keys', tablefmt='psql'))
```

```python
# Know mean of code challengue and technical enterview score
print('\nMean of Code Challenge Score:')
print(df['Code Challenge Score'].mean())

print('\nMean of Technical Interview Score:')
print(df['Technical Interview Score'].mean())

# Know
print('\nMode of Code Challenge Score:')
print(tabulate(df['Code Challenge Score'].value_counts().head(10).to_frame(), headers='keys', tablefmt='psql'))

print('\nMode of Technical Interview Score:')
print(tabulate(df['Technical Interview Score'].value_counts().head(10).to_frame(), headers='keys', tablefmt='psql'))
```
